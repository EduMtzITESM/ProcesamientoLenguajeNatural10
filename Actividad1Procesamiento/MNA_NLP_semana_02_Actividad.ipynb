*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!". 

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen. 

Indica cuántos resultados obtuviste.



```python
# Se crea una lista donde se pondrá las palabras que
# terminan con dos o más signos de exclamacion
palabras_con_exclamacion = []

# Se crea un vector con los índices de los renglones
# que contienen alguna palabra que termina con dos ó mas
# signos de admiración
indices_renglones_exclamacion = []

# Proceso iterativo para encontrar las palabras que 
# terminan con dos o más signos de admiración 
for oracion in docs_sin_saltos:
    palabras_minusc = oracion.lower()
    # se separa en strings individuales cada renglón
    palabras = palabras_minusc.split()
    for palabra in palabras:
        # se obtiene la condición de que la palabra termine con !!
        res = palabra.endswith("!!")
        if res:
            indice_oracion = docs_sin_saltos.index(oracion)
            indices_renglones_exclamacion.append(indice_oracion)
            indice_palabra = palabras.index(palabra)
            lista_interes = [palabras[indice_palabra].replace('!',''),
                             palabras[indice_palabra].count('!')]
            palabras_con_exclamacion.append(lista_interes)
```

Para mostrar las palabras (no necesariamente únicas) que terminan con al menos dos signos de admiración así como la totalidad de signos de admiración que le siguen:

```python
# Para contar el número de palabras que cumplen la condición
len(palabras_con_exclamacion)
```

```python
palabras_unicas_con_exclamacion = []

for lista in palabras_con_exclamacion:
    palabras_unicas_con_exclamacion.append(lista[0])

# Se obtiene el número de palabras únicas
len(set(palabras_unicas_con_exclamacion))
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
# Se crea una lista donde se pondrá las palabras que
# están escritas totalmente en mayúsculas
palabras_solo_mayusculas = []

# Se crea un vector con los índices de los renglones
# que contienen alguna palabra que termina con dos ó mas
# signos de admiración
indices_renglones_solo_mayusculas = []

for oracion in docs_sin_saltos:
    palabras = oracion.split()
    for palabra in palabras:
        palabra = palabra.strip()
        res = palabra.isupper()
        if res:
            indice_palabra = palabras.index(palabra)
            lista_interes = palabras[indice_palabra]
            palabras_solo_mayusculas.append(lista_interes)
            indice_oracion = docs_sin_saltos.index(oracion)
            indices_renglones_solo_mayusculas.append(indice_oracion)
            

palabras_solo_mayusculas_limpia = []

for palabra in palabras_solo_mayusculas:
    palabra = re.sub(r'[^\w\s]','',palabra)
    palabras_solo_mayusculas_limpia.append(palabra)
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú. 

Indica cuántos resultados obtuviste.

```python
vocal_acentuada = "ÁáÉéÍíÓóÚú"

# Se crea una lista donde se pondrá las palabras que
# están escritas totalmente en mayúsculas
palabras_vocales = []

# Se crea un vector con los índices de los renglones
# que contienen alguna palabra que termina con dos ó mas
# signos de admiración
indices_renglones_vocales = []

for oracion in docs_sin_saltos:
    palabras = oracion.split()
    for palabra in palabras:
        palabra = palabra.strip()
        vocs_acen = [x for x in palabra if x in vocal_acentuada]
        if len(vocs_acen):
            indice_palabra = palabras.index(palabra)
            lista_interes = palabras[indice_palabra]
            palabras_vocales.append(lista_interes)
            indice_oracion = docs_sin_saltos.index(oracion)
            indices_renglones_vocales.append(indice_oracion)
            
palabras_vocales_limpia = []

for palabra in palabras_vocales:
    palabra = re.sub(r'[^\w\s]','',palabra)
    palabras_vocales_limpia.append(palabra)
```

Busca e imprime todas las palabras que tengan una longitud mayor estrictamente a 10 caracteres alfabéticos.

No se consideran los signos de puntuación o caracteres especiales en la longitud de estas cadenas, solo caracteres alfabéticos en mayúsculas o minúsculas.

```python
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
palabras_long_10 = []

for oracion in docs_sin_saltos:
    palabras_minusc = oracion.lower()
    palabras_atom = re.findall( r'\w+|[^\s\w]+', palabras_minusc)
    for palabra in palabras_atom:
        if palabra in punc:
            palabra = palabra.replace(palabra, "")
        res = len(palabra) > 10
        if res:
            palabras_long_10.append(palabra)
```

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes. 
   




```python
lista_renglon_corpus = []

for oracion in docs_sin_saltos:
    indice_oracion = docs_sin_saltos.index(oracion)
    oracion = oracion.lower()
    palabras = oracion.split()
    renglon_limpio = []
    for palabra in palabras:
        palabra = palabra.strip()
        palabra = re.sub(r'[^\w\s]','',palabra)
        palabra = re.sub(r'[0-9]','',palabra)
        print(palabra)
        renglon_limpio.append(palabra)
    lista_renglon_corpus.append([indice_oracion, renglon_limpio])
```

```python
lista_comentarios_limpios = []

for i in range(len(docs_sin_saltos)):
    l = lista_renglon_corpus[i][1]
    l = ' '.join(l)
    lista_comentarios_limpios.append(l)
```

```python
for comentario in lista_comentarios_limpios[0:10]:
    print(comentario)
```

Las expresiones regulares son una herramienta muy poderosa y útil para trabajar con cadenas de texto, en particular cuando se empieza a estudiar NLP. las expresiones regulares son: patrones de coincidencia. Patrones que pueden usarse para comparar, extraer, remplazar o dividir segmentos de un texto particular, en otra cadena, texto largo o documento.

Las expresiones regulares son complicadas y de difícil retención ya que se escriben muy pocos símbolos reservados, casi todos compuestos por un carácter. De aquí que se requiere mucha concentración leer una regex que no se conoce, que no escribimos nosotros ó que no la escribimos hace mucho tiempo.

Aunque difícil, no quiere decir que aprender a escribir dichas expresiones regulares sea imposible o tan difícil. Con muy poco esfuerzo se puede aprender lo básico.  Por ejemplo, una expresión regular con literales como `/xyz/` va a coincidir la secuencia `xyz` en la cadena `ejemplo axyz dos`. Las barras son necesarias dependiendo de la herramienta/librería a la que le se le pasa la regex.
